```python
import pandas as pd

# -----------------------------
# Step 1: Create Sample Sales Data
# -----------------------------
data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Sales": [1200,1350,1400,1500,1650,1700,1800,1900,2000,2100,2250,2400]
}

df = pd.DataFrame(data)

print("Sales Dataset")
print(df)

# -----------------------------
# Step 2: Time-Based Data Analysis
# -----------------------------
print("\nAverage Sales:", round(df["Sales"].mean(),2))
print("Highest Sales:", df["Sales"].max())
print("Lowest Sales:", df["Sales"].min())

# -----------------------------
# Step 3: Trend Analysis
# -----------------------------
if df["Sales"].iloc[-1] > df["Sales"].iloc[0]:
    print("\nTrend Analysis")
    print("Sales show an increasing trend throughout the year.")
else:
    print("\nTrend Analysis")
    print("Sales are decreasing.")

# -----------------------------
# Step 4: Seasonality
# -----------------------------
print("\nSeasonality")
print("Sales are higher during the last quarter (Oct-Dec), indicating festive season demand.")

# -----------------------------
# Step 5: Forecasting
# -----------------------------
growth = df["Sales"].diff().mean()

next_month_sales = df["Sales"].iloc[-1] + growth

print("\nForecast for Next Month")
print("Expected Sales:", round(next_month_sales))

# -----------------------------
# Step 6: Business Impact Analysis
# -----------------------------
print("\nBusiness Impact Analysis")
print("1. Demand is increasing over time.")
print("2. Inventory should be increased before festive months.")
print("3. Marketing campaigns should focus on high-demand periods.")
print("4. Better demand forecasting helps reduce stock shortages.")

print("\nCapstone Project 3 Completed Successfully!")
```
